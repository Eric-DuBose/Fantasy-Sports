```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
skaters = pd.read_csv('KoH/all_players_koh.csv')
team = pd.read_csv('KoH/wild_koh.csv')
```

```python
def analysis(df1, df2, n):
    df1 = df1[df1['GP'] >= n]
    df1 = df1.replace(np.nan, '', regex=True)
    df2['My_Team'] = 'Y'
    df2 = df2.replace(np.nan, '', regex=True)
    both = pd.merge(df1, df2[['Player','My_Team']], left_on='Player', right_on='Player',how = 'left')
    both['My_Team'] = both['My_Team'].fillna('N')
    #both['C'] = np.where(both['Eligible'].str.contains('C'),'Y','N')
    #both['LW'] = np.where(both['Eligible'].str.contains('LW'),'Y','N')
    #both['RW'] = np.where(both['Eligible'].str.contains('RW'),'Y','N')
    #both['D'] = np.where(both['Eligible'].str.contains('D'),'Y','N')
    both[['Salary','G', 'A', 'PIM', 'SOG', 'GWG', 'PPP', 'SHP', 'Hit','Blk', 'FOW']] = round(both[['Salary','G', 'A', 'PIM', 'SOG', 'GWG', 'PPP', 'SHP', 'Hit','Blk', 'FOW']].div(both.GP, axis = 0),2)
    columns = ['Player', 'Team','KoH Team', 'My_Team', 'Salary','G',
       'A', 'PIM', 'SOG', 'GWG', 'PPP', 'SHP', 'Hit', 'Blk', 'FOW']
    return both[columns]

def stat_rank(data):
    data = data.drop('Salary', axis = 1)
    data[['G', 'A', 'PIM', 'SOG', 'GWG', 'PPP', 'SHP', 'Hit','Blk', 'FOW']] = data[['G', 'A', 'PIM', 'SOG', 'GWG', 'PPP', 'SHP', 'Hit','Blk', 'FOW']].rank(ascending = False).astype(int)
    return data

def salary_rank(data):
    data = data.copy()
    data[['G', 'A', 'PIM', 'SOG', 'GWG', 'PPP', 'SHP', 'Hit','Blk', 'FOW']] = data[['G', 'A', 'PIM', 'SOG', 'GWG', 'PPP', 'SHP', 'Hit','Blk', 'FOW']].div(1/data.Salary, axis = 0).astype(int)
    data = data.drop('Salary', axis = 1)
    data[['G', 'A', 'PIM', 'SOG', 'GWG', 'PPP', 'SHP', 'Hit','Blk', 'FOW']] = data[['G', 'A', 'PIM', 'SOG', 'GWG', 'PPP', 'SHP', 'Hit','Blk', 'FOW']].rank(ascending = False).astype(int)
    return data
```

```python
analysis = analysis(skaters, team,10)
stats = stat_rank(analysis)
salary = salary_rank(analysis)
```

```python
stats = stats[stats['KoH Team'] != 'FA']
stats.groupby('KoH Team').sum()
```
