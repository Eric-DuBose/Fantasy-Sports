```python
import pandas as pd
import numpy as np
```

```python
zips_batters = pd.read_csv('zips_batters.csv')
zips_pitchers = pd.read_csv('zips_pitchers.csv')
zips_breakdown = pd.concat([zips_batters,zips_pitchers], ignore_index = True, sort = False).fillna(0)
tbrl_batters = pd.read_csv('tbrl_salaries.csv')
pitchers = zips_pitchers[['Name','Team','POS','Dollars']]
batters = zips_batters[['Name','Team','POS','Dollars']]
zips_all = pd.concat([pitchers,batters])
zips_all = zips_all.rename(columns = {'Team':'ML Team'})
fa_pool = pd.concat([zips_batters, zips_pitchers]).fillna('').set_index('Name')
```

```python
def pos_chart(pos):
    data = zips_breakdown.copy()
    data = data[data['Dollars'] > .1]
    data = data[['Name','Team', 'POS', 'zAVG', 'zRBI', 'zSB', 'zHR','zW', 'zSV', 'zERA', 'zWHIP', 'zTOT','Dollars']]
    return data[data['POS'].str.contains(pos)]

def projections():
    projections = tbrl_batters.merge(zips_all, on = 'Name')
    projections['Surplus'] = projections['Dollars'] - projections['Salary'].astype(float)
    return projections[['Team', 'Name', 'POS', 'ML Team', 'Salary','Dollars','Surplus']]

tbrl_teams = projections()
```
